```python
import pandas as pd
import matplotlib as plt
import numpy as np
```

Grupo 1

```python
#cargar el conjunto de datos
grupo1 = pd.read_csv('grupo1.csv', encoding ='ISO-8859-1')
grupo1.head()
```

```python
#descripcion estadistica
grupo1.describe()
```

```python
#visualizacin
import matplotlib.pyplot as plt
grupo1['Calificacion'].hist()
plt.title('Histograma de calificaciones')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# entrenar el modelo
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
ln = LinearRegression()
x = grupo1[['HorasEstudio']]
y = grupo1['Calificacion']
x.head()
```

```python
ln.fit(x,y)
```

```python
Ypred = ln.predict(x)
```

```python
print("Valor de la pendiente:", ln.coef_)
```

```python
print("Valor de la intercepcin: ", ln.intercept_)
```

```python
#diagrama de dispersin con lnea de regresin
import matplotlib.pyplot as plt
plt.figure(figsize=(6,5))
plt.scatter(x, y, color = 'blue', label='datos')
plt.plot(x, ln.predict(x), color = 'red', label='linea de regresin')
plt.scatter(x, Ypred, color = 'green', marker='x', s=100,
            label='prediccin (8, {0:.2f})'.format(Ypred[0]))
plt.title("Horas de estudio vs calificaciones")
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(6, 4))
sns.set(style = "whitegrid")
sns.regplot(x=x, y=y, scatter=True, color='blue')
plt.title("Horas de estudio vs calificaciones")
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.grid(True)
plt.show()
```

```python
#coeficiente de pearson y valor de p
from scipy.stats import pearsonr
corr, valor_p = pearsonr(grupo1['HorasEstudio'], grupo1['Calificacion'])
print('Coeficiente de correlacin:', corr)
print('Valor de p: ', valor_p)
```

####**Interpretacin de resultados**


*   **Cul es la interpretacin del coeficiente de regresin en trminos de la
relacin entre el tiempo de estudio y las calificaciones finales?** El coeficiente de regresin es aproximadamente de -0.11. Esto significa que por cada hora de estudio, la calificacin final disminuye en 0.11 puntos, aproximadamente. Esta relacin sugiere deficiencia en el modelo o en los datos.
*   **Cmo se interpreta el coeficiente de correlacin en trminos de la fuerza y direccin de la relacin entre las variables?** El coeficiente de correlacin es aproximadamente de -0.06, y el valor p es 0.69. Esto indica una correlacin muy debil y negativa. La probabilidad de obtener este resultado si no existe la relacin es alta.
*   **El modelo de regresin lineal es estadsticamente significativo?** El valor p (0.69) es mayor a 0.05, indicando que el modelo de regresin no es estadsticamente significativo.
*   **Qu tan confiable es el modelo de regresin para predecir las
calificaciones finales de los estudiantes basado en el tiempo de estudio?** La correlacin dbil sugiere un modelo poco confiable para predecir calificaciones. El diagrama de dispersin mostrara una gran dispersin de los puntos.
