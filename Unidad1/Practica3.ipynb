```python
import random
# libreria para generar gráficos
import matplotlib.pyplot as plt
# libreria para usar funciones de tipo estadística
from statistics import *
```

```python
# crear la primera fábrica
fabrica1 = []
for i in range(100):
  num_aleatorio = round(random.uniform(50,100), 2)
  fabrica1.append(num_aleatorio)
print(fabrica1)
```

```python
# crear la segunda fábrica
fabrica2 = []
for i in range(100):
  num_aleatorio = round(random.uniform(50,100), 2)
  fabrica2.append(num_aleatorio)
print(fabrica2)
```

#### Máximo, mínimo y rango

```python
maximo1 = max(fabrica1)
minimo1 = min(fabrica1)
rango1 = maximo1 - minimo1
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)
print("Estadística descriptiva de la fábrica 1 \n", "Media:", media1,
      "\n Mediana:", mediana1, "\n Moda: ", moda1, "\n Máximo:", maximo1,
      "\n Mínimo:", minimo1, "\n Rango:", rango1)
```

```python
maximo2 = max(fabrica2)
minimo2 = min(fabrica2)
rango2 = maximo2- minimo2
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)
print("Estadística descriptiva de la fábrica 2 \n", "Media:", media2,
      "\n Mediana:", mediana2, "\n Moda: ", moda2, "\n Máximo:", maximo2,
      "\n Mínimo:", minimo2, "\n Rango:", rango2)
```

###Dispersión de datos

```python
dve_fabrica1 = stdev(fabrica1)
print("Desviación estándar de la fábrica 1:", dve_fabrica1)
```

```python
dve_fabrica2 = stdev(fabrica2)
print("Desviación estándar de la fábrica 2:", dve_fabrica2)
```

```python
plt.figure(figsize=(8,6))
plt.hist(fabrica1, bins=10, alpha=0.2, color='b', edgecolor ='black')
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label=f'Media={media1:.2f}')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana1:.2f}')
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label=f'Moda={moda1:.2f}')
plt.axvline(media1 - dve_fabrica1, color='black', linestyle='dashed', linewidth=2, label=f'Desviación estándar={dve_fabrica1:.2f}')
plt.axvline(media1 + dve_fabrica1, color='black', linestyle='dashed')

plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Producción diaria de la fábrica 1')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(fabrica2, bins=10, alpha=0.2, color='b', edgecolor ='black')
plt.axvline(media2, color='b', linestyle='dashed', linewidth=2, label=f'Media={media2:.2f}')
plt.axvline(mediana2, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana2:.2f}')
plt.axvline(moda2, color='y', linestyle='dashed', linewidth=2, label=f'Moda={moda2:.2f}')
plt.axvline(media2 - dve_fabrica2, color='black', linestyle='dashed', linewidth=2, label=f'Desviación estándar={dve_fabrica2:.2f}')
plt.axvline(media2 + dve_fabrica2, color='black', linestyle='dashed')

plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Producción diaria de la fábrica 2')
plt.legend()
plt.show()
```

###Densidad

```python
import seaborn as sns
```

```python
# la densidad gráfica es una forma de visualizar la distribución de datos en
# un intérvalo o periodo de tiempo continuo.
# Se basa en un histograma que usa el suavizado de kernel pars trazar valores,
# lo que permite obtener distribuciones más suaves al reducir el ruido.
# Una curva de densiddad muestra la ubicación de la media y la mediana de una distribución.
sns.histplot(x = fabrica1, kde = True, kde_kws= {'bw_adjust': 0.5}, bins = 10)
```

```python
sns.histplot(x = fabrica2, kde = True, kde_kws= {'bw_adjust': 0.5}, bins = 10)
```
