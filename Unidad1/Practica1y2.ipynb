```python
from statistics import *
# crear lista
g = [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100,96]
print(g)
```

```python
type(g)
```

```python
# calcular la media
prom = sum(g) / len(g)
print(prom)
```

```python
# funci√≥n de la librer√≠a statistics para calcular la media
prom2 = mean(g)
print(prom2)
```

```python
# calcular la mediana con la funci√≥n med
med = median(g)
print(med)
```

```python
#calcular la mediana con la funci√≥n mode
moda = mode(g)
print(moda)
```

```python
# se utiliza la funci√≥n multimode en caso de que haya m√°s de una moda
moda2 = multimode(g)
print(moda2)
```

```python
# tiempos de 10 alumnos que correr√°n 10 metros
tiempos = [15.10, 17.20, 14.69, 13.27, 22.15, 18.71, 19.15, 20.65, 15.10, 17.20]
print(tiempos)
```

```python
# determinar el tiempo promedio
print("El tiempo promedio de los corredores es:", mean(tiempos))
```

```python
# sacar la mediana
print("La media de los tiempos de los corredores es:", median(tiempos))
```

```python
# determinar la moda
print("La moda de los tiempos de los corredores es:", mode(tiempos))
```

```python
# determinar las multiples modas
print("Las modas de los tiempos son:", multimode(tiempos))
```

```python
# Librer√≠a para generar numeros aleatorios
import random
```

```python
# generar los numeros aleatorios con el ciclo for
fabrica1 = []
for i in range (100):
  num_aleatorio1 = round( random.uniform(50, 100),2)
  fabrica1.append(num_aleatorio1)
print(fabrica1)
```

```python
fabrica2 = []
for i in range (100):
  num_aleatorio2 = round( random.uniform(50, 100),2)
  fabrica2.append(num_aleatorio2)
print(fabrica2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Producci√≥n diaria de f√°brica 1")
```

```python
plt.hist(fabrica2)
plt.title("Producci√≥n diaria de f√°brica 2")
```

```python
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)
print("Valores de la fabrica 1")
print("La media es :", media1)
print("La mediana es : ", mediana1)
print("La moda es :", moda1)
```

```python
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)
print("Valores de la fabrica 2")
print("La media es :", media2)
print("La mediana es : ", mediana2)
print("La moda es :", moda2)
```

```python
fig, (ax1, ax2)=plt.subplots(nrows =1, ncols=2, figsize= (10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

```python
plt.hist(fabrica1, bins=10, alpha=0.2, color='b', edgecolor = "black")
plt.axvline(media1, color='b', linestyle= 'dashed', linewidth=2, label = 'Media')
plt.axvline(mediana1, color='r', linestyle= 'dashed', linewidth=2, label = 'Mediana')
plt.axvline(moda1, color='y', linestyle= 'dashed', linewidth=2, label = 'Moda')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Produccion diaria de la fabrica 1')
plt.legend()
plt.show()
```

```python
plt.hist(fabrica2, bins=10, alpha=0.2, color='b', edgecolor = "black")
plt.axvline(media2, color='b', linestyle= 'dashed', linewidth=2, label = 'Media')
plt.axvline(mediana2, color='r', linestyle= 'dashed', linewidth=2, label = 'Mediana')
plt.axvline(moda2, color='y', linestyle= 'dashed', linewidth=2, label = 'Moda')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Produccion diaria de la fabrica 2')
plt.legend()
plt.show()
```
