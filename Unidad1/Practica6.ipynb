```python
# importar las librerias
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
autos = pd.read_csv('autos2 (1).csv', encoding ='ISO-8859-1')
```

```python
autos.head(10)
```

```python
autos.info()
```

```python
# obtener el valor mximo de los precios
autos.price.max()
```

```python
# obtener el valor mnimo de los precios
autos.price.min()
```

```python
# determinar la media
media = autos.price.mean()
print(media)
```

```python
#determinar la mediana
mediana = autos.price.median()
print(mediana)
```

```python
#determinar la moda
moda = autos.price.mode()
print(moda)
```

```python
#determinar la desviacin estndar de los precios
desviacion = autos.price.std()
print(desviacion)
```

```python
#determinar los estadsticos y cuartiles
autos.price.describe()
```

```python
#calcular el numero de intervalos con la regla de Sturges
import math
n = len(autos.price)
K = int(1 + math.log2(n))
print("Valor de n:", n, "\nNmero de intrvalos:", K)
```

```python
# crear el histograma de los precios
plt.pyplot.hist(autos.price, bins=K)
plt.pyplot.title('Histograma de precios de los automviles')
plt.pyplot.axvline(media, color='y', linestyle= 'dashed', linewidth=2, label = 'Media')
plt.pyplot.axvline(mediana, color='r', linestyle= 'dashed', linewidth=2, label = 'Mediana')
#no se puede mostrar la moda debido a que aparece en una lista
plt.pyplot.xlabel('Precio')
plt.pyplot.ylabel('Frecuencia')
plt.pyplot.legend()
```

Este histograma muestra que hay aproximadamente 100 autos que rondan el precio de 5000 a 10000, mientras que del precio de 10000 a 15000 hay aproximadamente 40 autos, de 15000 a 20000 hay un poco ms de 30 autos, y en el resto de precios hay aproximadamente 10 autos o menos; esto indica que hay ms autos de precio bajo que de precio alto.

```python
#relacin estadstica entre el precio y tipo de vehculo
sns.boxplot(x='body-style', y='price',data=autos)
plt.pyplot.title('Relacin entre el precio y el tipo de vehculo')
```

En este diagrama, se muestra que los convertibles son los ms caros, mientras que los hatchback y sedan son los ms baratos. Los precios de los wagon y hardtop se encuentran en un rango intermedio. Algunos precios individuales se alejan significativamente del resto, por lo que son valores atpicos.

```python
#relacin estadstica entre el precio y tipo traccin
sns.boxplot(x='drive-wheels', y='price', data=autos)
plt.pyplot.title('Relacin entre el precio y el tipo de ruedas')
```

Los vehculos de traccin rwd tienden a ser ms caros que los de traccin fwd y los de traccin 4wd. Hay algunos precios atpicos en cada categora, indicando la existencia de vehculos con precios inusualmente altos o bajos dentro de cada tipo de traccin.

```python
#relacin estadstica entre el precio y la ubicacin del motor
sns.boxplot(x='engine-location', y='price', data=autos)
plt.pyplot.title('Relacin entre el precio y la ubicacin del motor')
```

Los autos con motor trasero son significativamente ms caros que los autos con motor delantero. La mediana del precio es mucho mayor para los autos con motor trasero, y hay una mayor variabilidad en los precios de los autos con motor delantero, como lo indican los bigotes ms largos. Tambin hay valores atpicos en los autos con motor delantero que indican precios inusualmente altos.
