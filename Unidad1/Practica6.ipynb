```python
# importar las librerias
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
autos = pd.read_csv('autos2 (1).csv', encoding ='ISO-8859-1')
```

```python
autos.head(10)
```

```python
autos.info()
```

```python
# obtener el valor máximo de los precios
autos.price.max()
```

```python
# obtener el valor mínimo de los precios
autos.price.min()
```

```python
# determinar la media
media = autos.price.mean()
print(media)
```

```python
#determinar la mediana
mediana = autos.price.median()
print(mediana)
```

```python
#determinar la moda
moda = autos.price.mode()
print(moda)
```

```python
#determinar la desviación estándar de los precios
desviacion = autos.price.std()
print(desviacion)
```

```python
#determinar los estadísticos y cuartiles
autos.price.describe()
```

```python
#calcular el numero de intervalos con la regla de Sturges
import math
n = len(autos.price)
K = int(1 + math.log2(n))
print("Valor de n:", n, "\nNúmero de intérvalos:", K)
```

```python
# crear el histograma de los precios
plt.pyplot.hist(autos.price, bins=K)
plt.pyplot.title('Histograma de precios de los automóviles')
plt.pyplot.axvline(media, color='y', linestyle= 'dashed', linewidth=2, label = 'Media')
plt.pyplot.axvline(mediana, color='r', linestyle= 'dashed', linewidth=2, label = 'Mediana')
#no se puede mostrar la moda debido a que aparece en una lista
plt.pyplot.xlabel('Precio')
plt.pyplot.ylabel('Frecuencia')
plt.pyplot.legend()
```

Este histograma muestra que hay aproximadamente 100 autos que rondan el precio de 5000 a 10000, mientras que del precio de 10000 a 15000 hay aproximadamente 40 autos, de 15000 a 20000 hay un poco más de 30 autos, y en el resto de precios hay aproximadamente 10 autos o menos; esto indica que hay más autos de precio bajo que de precio alto.

```python
#relación estadística entre el precio y tipo de vehículo
sns.boxplot(x='body-style', y='price',data=autos)
plt.pyplot.title('Relación entre el precio y el tipo de vehículo')
```

En este diagrama, se muestra que los convertibles son los más caros, mientras que los hatchback y sedan son los más baratos. Los precios de los wagon y hardtop se encuentran en un rango intermedio. Algunos precios individuales se alejan significativamente del resto, por lo que son valores atípicos.

```python
#relación estadística entre el precio y tipo tracción
sns.boxplot(x='drive-wheels', y='price', data=autos)
plt.pyplot.title('Relación entre el precio y el tipo de ruedas')
```

Los vehículos de tracción rwd tienden a ser más caros que los de tracción fwd y los de tracción 4wd. Hay algunos precios atípicos en cada categoría, indicando la existencia de vehículos con precios inusualmente altos o bajos dentro de cada tipo de tracción.

```python
#relación estadística entre el precio y la ubicación del motor
sns.boxplot(x='engine-location', y='price', data=autos)
plt.pyplot.title('Relación entre el precio y la ubicación del motor')
```

Los autos con motor trasero son significativamente más caros que los autos con motor delantero. La mediana del precio es mucho mayor para los autos con motor trasero, y hay una mayor variabilidad en los precios de los autos con motor delantero, como lo indican los bigotes más largos. También hay valores atípicos en los autos con motor delantero que indican precios inusualmente altos.
