```python
import random
```

```python
edades = []
for i in range(100):
  #generar un numero aleatorio con la funcin randit
  num_aleatorio = random.randint(18, 70)
  #agregar el numero aleatorio a la lista de edades
  edades.append(num_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
#crear un DataFrame de pandas con los numeros aleatorios
df = pd.DataFrame(edades, columns=["Edad"])
df.head(10)
```

```python
#calcular la tabla de distribucin de frecuencias
tabla_frecuencias = df["Edad"].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')
#imprimir la tabla
print("Tabla de distribucin de frecuencias:")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec relativa'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec porcentual'] = tabla_frecuencias['Frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

###Regla de Sturges

##### Regla que sirve para calcular la cantidad ideal de clases o intrvalos en las que se debe dividir un conjunto de datos

```python
import math
```

```python
# volver a imprimir la lista de edades
print(edades)
```

```python
df = pd.DataFrame(edades, columns=["Edad"])
df
```

```python
#calcular el numero de intervalos
n = len(edades)
K = int(1 + math.log2(n))
print("Valor de n:", n, "\nNmero de intrvalos:", K)
```

```python
tabla_frecuencias = pd.cut(df['Edad'], bins = K, include_lowest = True).value_counts().reset_index()
tabla_frecuencias.columns = ['Clase', 'Frec']
tabla_frecuencias
```

```python
# ordenar la tabla por frecuencia de intervalo
tabla_frecuencias = tabla_frecuencias.sort_values(by = 'Clase')
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Rel'] = tabla_frecuencias['Frec'] / n
tabla_frecuencias['Frec Porc'] = tabla_frecuencias['Frec'] / n * 100
tabla_frecuencias['Frec Ac'] = tabla_frecuencias['Frec'].cumsum()
tabla_frecuencias['Frec Rel Ac'] = tabla_frecuencias['Frec Rel'].cumsum()
tabla_frecuencias['Frec Porc Ac'] = tabla_frecuencias['Frec Porc'].cumsum()
print(tabla_frecuencias)
```

```python
import matplotlib.pyplot as plt
plt.hist(edades, alpha = 0.5, bins = 15)
plt.title("Histograma de edades basado en la funcin de matplotlib con valores estndar")
```

```python
#histograma basado en la tabla de frecuencias
plt.hist(df['Edad'], bins = K, alpha = 0.5)
plt.title("Histograma de edades basado en la regla de Sturges")
plt.xlabel("Edades")
plt.ylabel("Frecuencia")
```

##Prctica 5: Diagramas de cajas y bigotes


```python
edades2 = []
for i in range(100):
  #generar un numero aleatorio con la funcin randit
  num_aleatorio = random.randint(20, 87)
  #agregar el numero aleatorio a la lista de edades
  edades2.append(num_aleatorio)
print(edades2)
```

```python
#crear el diagrama para comparar las dos encuestas
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (9,4))
bplot1 = ax1.boxplot(edades,
                     vert = True,  #vertical box alignment
                     patch_artist = True) #fill with color
ax1.set_title("Edades encuesta 1")
bplot2 = ax2.boxplot(edades2,
                     vert = True,  #vertical box alignment
                     patch_artist = True) #fill with color
ax2.set_title("Edades encuesta 2")
plt.show()
```
