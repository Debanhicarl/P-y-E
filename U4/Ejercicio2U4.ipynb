```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

Probabilidad de que un estudiante pese menos de 75 kg

```python
plt.figure(figsize=(6, 4))
#parametros de la distribución normal
media = 70
desv_std = 4
total_estud = 500
#calcculo de probabilidad que un estudiante pese menos de 75kg
peso_lim = 75
prob_peso = norm.cdf(peso_lim,media,desv_std)
#calculo del num de estuiantes que pesen menos de 75 kg
estud_menos_75kg = total_estud * prob_peso
print("Probabilidad de que un estudiante pese menos de 75 kg:", prob_peso)
print("Número de estudiantes que pesen menos de 75 kg:", estud_menos_75kg)
#graficar
x = np.linspace(media - 4*desv_std, media + 4*desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_lim), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75 kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

Probabilidad de que un estudiante pese más de 79 kg

```python
plt.figure(figsize=(6, 4))
#parametros de la distribución normal
media = 70
desv_std = 4
total_estud = 500
#calcculo de probabilidad que un estudiante pese más de 79kg
peso_lim = 79
prob_peso_menos= norm.cdf(peso_lim,media,desv_std)
prob_mas = 1 - prob_peso_menos
#calculo del num de estuiantes que pesen más de 79 kg
estud_mas_79kg = total_estud * prob_mas
print("Probabilidad de que un estudiante pese más de 79 kg:", prob_mas)
print("Número de estudiantes que pesen más de 79 kg:", estud_mas_79kg)
#graficar
x = np.linspace(media - 4*desv_std, media + 4*desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_lim), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese más de 79 kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
