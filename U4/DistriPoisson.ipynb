Ejercicio 1

```python
from scipy.stats import poisson
# parametros de la distribución de poisson
media_accidentes = 2
# parte 1: probabilidad de que no haya accidentes en una semana
probabilidad_sin_accidentes = poisson.pmf(0, media_accidentes)
print("1. Probabilidad de que no haya accidentes en una semana: ", probabilidad_sin_accidentes)
# parte 2: probabilidad de que a lo sumo haya tres accidentes en 2 semanas
#convertir la media a una media para dos semanas
media_dos_semanas = media_accidentes * 2
#calculamos la probabilidad acumulada hasta 3 accidentes en 2 semanas
probabilidad_hasta_tres_accidentes = poisson.cdf(3, media_dos_semanas)
print("2. Probabilidad de que haya 3 accidentes en dos semanas: ", probabilidad_hasta_tres_accidentes)
```

```python
#parametros de la distribución de poisson
media_muertes = 5000 * 0.001 #5000 hombres con una probabilidad de muerte de 0.001
#numero de reclamaciones
x = 4
#calcular la probabilidad de que la compañia tenga que pagar 4 reclamaciones en un año determinado
probabilidad_reclamaciones = poisson.pmf(x, media_muertes)
print("La probabilidad de que la compañia tenga que pagar 4 reclamaciones en un año determinado es: ", probabilidad_reclamaciones)
```
